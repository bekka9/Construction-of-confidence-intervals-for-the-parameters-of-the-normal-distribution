## Построение доверительных интервалов для параметров нормального распределения



```python
import pandas as pd
import statistics
import numpy as np
import pandas
import scipy.stats as sts
import statistics as stat

df = pd.DataFrame(pd.read_excel("/content/mat_stat2(19_2).xlsx"))
n = 50
q = 0.95
eps = 1 - 0.95
df
```

```python
x1 = df.sort_values(['X1'])['X1']
x1.reset_index(drop= True , inplace= True )
x2 = df.sort_values(['X2'])['X2']
x2.reset_index(drop= True , inplace= True )
x3 = df.sort_values(['X3'])['X3']
x3.reset_index(drop= True , inplace= True )
```

## 1. Доверительный интервал для математического ожидания *a*

```python
var1 = 16
sd1 = np.sqrt(var1) #Среднеквадратическое отклонение x1
x_mean = stat.mean(x1)

#доверительные пределы при известном среднеквадратическом отклонении (x1):
#left1 = x_mean - np.quantile(x1, 1 - eps/2) * sd1/(np.sqrt(n))
left1 = x_mean - sts.norm.ppf(1 - eps/2) * sd1/(np.sqrt(n))
right1 = x_mean + sts.norm.ppf(1 - eps/2) * sd1/(np.sqrt(n))
print('(', left1, ';', right1, ')\n')

#доверительные пределы при неизвестном среднеквадратическом отклонении (x2):
left2 = stat.mean(x2) - sts.t.ppf(1 - eps/2, n-1) * stat.stdev(x2)/(np.sqrt(n - 1))
right2 = stat.mean(x2) + sts.t.ppf(1 - eps/2, n-1) * stat.stdev(x2)/(np.sqrt(n - 1))
print('(', left2, ';', right2, ')\n')
```

# 2.	Доверительный интервал для дисперсии   (для выборки Х3):

```python
a3 = 10 #математическое ожидание x3
mean3 = stat.mean(x3)
#доверительные пределы для дисперсии при известном мат.ожидании (x3):
left3 = n * (stat.mean(x3**2)- 2*a3*mean3 + a3**2) / sts.chi2.ppf(1 - eps/2, n)
right3 = n * (stat.mean(x3**2)- 2*a3*mean3 + a3**2) / sts.chi2.ppf(eps/2, n)
print('(', left3, ';', right3, ')\n')

#доверительные пределы при неизвестном мат.ожидании (x2):
left_2 = n * (stat.stdev(x2)**2)/ sts.chi2.ppf(1 - eps/2, n - 1)
right_2 = n * (stat.stdev(x2)**2) / sts.chi2.ppf(eps/2, n - 1)
print('(', left_2, ';', right_2, ')\n')
```
